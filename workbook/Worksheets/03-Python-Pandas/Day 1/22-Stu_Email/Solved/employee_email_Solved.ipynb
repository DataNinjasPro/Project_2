
<br>
import os<br>
import csv<br>
filepath = os.path.join("Resources", "employees.csv")<br>
new_employee_data = []<br>
# Read data into dictionary and create a new email field<br>
with open(filepath) as csvfile:<br>
    reader = csv.DictReader(csvfile)<br>
    for row in reader:<br>
        first_name = row["first_name"]<br>
        last_name = row["last_name"]<br>
        email = f"{first_name}.{last_name}@example.com"<br>
        new_employee_data.append(<br>
            {<br>
                "first_name": row["first_name"],<br>
                "last_name": row["last_name"],<br>
                "ssn": row["ssn"],<br>
                "email": email<br>
            }<br>
        )<br>
# Grab the filename from the original path<br>
_, filename = os.path.split(filepath)<br>
# Write updated data to csv file<br>
csvpath = os.path.join("output", filename)<br>
with open(csvpath, "w") as csvfile:<br>
    fieldnames = ["last_name", "first_name", "ssn", "email"]<br>
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)<br>
    writer.writeheader()<br>
    writer.writerows(new_employee_data)
