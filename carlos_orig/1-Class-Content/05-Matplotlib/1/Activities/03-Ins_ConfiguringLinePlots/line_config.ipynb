```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set x axis and variables
x = np.arange(0, 10, 0.1)
sin = np.sin(x)
cos = np.cos(x)
```

```python
# Draw a horizontal line with 0.25 transparency
plt.hlines(0, 0, 10, alpha=0.25)
```

```python
# Note: The plot method returns a tuple of lines that were added to the plot. 
# The code `sine_handle,` uses argument unpacking to select the handle to the line object
# Otherwise, you would have get `[<matplotlib.lines.Line2D at 0x10cbf96d8>]`

# Each point on the sine chart is marked by a blue circle
sine_handle, = plt.plot(x, sin, marker='o', color='blue', label="Sine")
# Each point on the cosine chart is marked by a red triangle
cosine_handle, = plt.plot(x, cos, marker='^', color='red', label="Cosine")
```

```python
# Adds a legend and sets its location to the lower right
plt.legend(handles=[sine_handle, cosine_handle], loc="lower right")
```

```python
# Saves an image of our chart so that we can view it in a folder
plt.savefig("lineConfig.png")
plt.show()
```
